```
import numpy as np
import pandas as pd
import matplotlib
train = pd.read_csv("/work/Data/task3/train.csv")
train
```

```
df_1 = train
del df_1['Name']
del df_1 ['Sex']
del df_1 ['Ticket']
del df_1 ['Cabin']
del df_1 ['Embarked']
df_1
```

```
df_1 = df_1.dropna ()
df_1
```

```
X = df_1.loc[:,df_1.columns!="Survived"]
y = df_1['Survived']
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit (X_train,y_train)
y_pred = model_1.predict (X_test)
```

```
np.square(np.subtract(y_test,y_pred)).mean()
```

```
train_2 = pd.read_csv("/work/Data/task3/train_2.csv")
train_2
```

```
df_2 = train_2
df_2.drop(['Name','PassengerId','Ticket','Cabin','Embarked'], axis=1, inplace=True)
df_2 = df_2.dropna()
df_2
```

Мужчины и женщины по-разному переносят холод и могут держаться на воде. Класс нужен для того, чтобы понять кого спасали впервую очередь. Сайблингс он борд нужен, ибо ты же пойдешь спасать своих родственников скорее всего и вохможно умрешь из-за этого

```
from sklearn.preprocessing import LabelEncoder
df_2['Sex'] = LabelEncoder().fit_transform(df_2['Sex'])
```

```
df_2
```

```
corr = df_2.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```
X = df_2.loc[:,df_2.columns!="Survived"]
y = df_2['Survived']
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model_2 = LinearRegression()
model_2.fit (X_train,y_train)
y_pred = model_2.predict (X_test)
print(mean_squared_error(y_test,y_pred))
```
